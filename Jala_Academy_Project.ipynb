```python
# Q9. Concatenate two arrays along a new axis
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5])

# Pad y to match length of x
if x.shape[0] > y.shape[0]:
    y = np.pad(y, (0, x.shape[0] - y.shape[0]), mode='constant')
else:
    x = np.pad(x, (0, y.shape[0] - x.shape[0]), mode='constant')

# Now stack along a new axis
stacked = np.stack((x, y), axis=0)
print("Concatenated along new axis:\n", stacked)
```

```python
# Q29. Solve system: 2x + 3y = 5, 4x + 6y = 10
coeff = np.array([[2, 3], [4, 6]])
const = np.array([5, 10])
try:
    solution = np.linalg.solve(coeff, const)
    print("Solution:", solution)
except np.linalg.LinAlgError:
    print("No unique solution (singular matrix)")
```

```python
# Q30. Perform SVD on 3x3 matrix and reconstruct
mat8 = np.random.randint(1, 10, (3, 3))
U, S, VT = np.linalg.svd(mat8)
S_matrix = np.zeros((3, 3))
np.fill_diagonal(S_matrix, S)
reconstructed = np.dot(U, np.dot(S_matrix, VT))
print("Original Matrix:\n", mat8)
print("Reconstructed Matrix from SVD:\n", reconstructed)
```

```python
# Q58. Rolling mean on time series and plot
import matplotlib.pyplot as plt

df_time = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Value': [10, 12, 14, 16, 15, 18, 20, 21, 19, 22]
})
df_time.set_index('Date', inplace=True)
df_time['Rolling_Mean'] = df_time['Value'].rolling(window=3).mean()
df_time[['Value', 'Rolling_Mean']].plot(title='Rolling Mean of Value')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```
